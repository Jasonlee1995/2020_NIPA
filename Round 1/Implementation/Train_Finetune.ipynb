```python
from dataset import *
from train import *

import matplotlib.pyplot as plt

import torch
import torchvision
import torchvision.transforms as transforms
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
batch_size = 32

epochs = 15
lr = 0.00001
weight_decay = 0.000001
```

```python
train_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.4633, 0.4887, 0.4060], std=[0.1824, 0.1601, 0.1958])])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.4633, 0.4887, 0.4060], std=[0.1824, 0.1601, 0.1958])])


train_dataset = NIPA_Dataset(data_type='train', transform=train_transform)
val_dataset = NIPA_Dataset(data_type='val', transform=val_transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

```python
model = Baseline(model='resnet18', num_classes=train_dataset.n_labels, save='./best_single.pt')
```

```python
model.model.load_state_dict(torch.load('./best_single.pt'))
```

```python
model.train(train_loader, val_loader, epochs=epochs, lr=lr, weight_decay=weight_decay)
```

```python
label_fontsize = 25

plt.figure(figsize=(20, 10))
train_lossline, = plt.plot(model.train_losses, label='Train')
test_lossline, = plt.plot(model.test_losses, color='red', label='Test')
plt.legend(handles=[train_lossline, test_lossline], fontsize=20)
plt.xlabel('Step', fontsize=label_fontsize)
plt.ylabel('Loss', fontsize=label_fontsize)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
train_accline, = plt.plot(model.train_acc, label='Train')
test_accline, = plt.plot(model.test_acc, color='red', label='Test')
plt.legend(handles=[train_accline, test_accline], fontsize=20)
plt.xlabel('Step', fontsize=label_fontsize)
plt.ylabel('Acc', fontsize=label_fontsize)
plt.show()
```
